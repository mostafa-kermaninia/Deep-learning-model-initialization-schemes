### **glorot initialization scheme**

We initialized the biases to be 0 and the weights $W_{ij}$ at each layer with the below formula:
$$ W_{ij} \sim U[(-\frac {\sqrt{6}}{\sqrt{n_{j}+n_{j+1}}}),(\frac {\sqrt{6}}{\sqrt{n_{j}+n_{j+1}}})]$$

* $U[−a, a]$ is the uniform distribution in the interval $(−a, a)$
* $n_{j}$ is the number of units (neurons) of the previous layer and $n_{j+1}$ is the number of units in the current layer
* $W_{i,j}$ is every weight of each layer




### **In paper1** we saw **Soft-sign, sigmoid and hyperbolic tangent** activation functions:

### **Sigmoid function (Logistic function):**
$$ f(x) = \frac{1}{1 + e^{-x}}$$
$$$$
### **Hyperbolic tangent (tanh) function:**
$$ f(x) = \tanh(x) = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}} $$
$$ $$

### **Softsign function:**
$$ f(x) = \frac{x}{1 + |x|} $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the hyperbolic tangent function (tanh)
def tanh(x):
    return np.tanh(x)

# Define the softsign function
def softsign(x):
    return x / (1 + np.abs(x))

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate y values using the sigmoid, tanh, and softsign functions
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_softsign = softsign(x)

# Plot all three functions
plt.figure(figsize=(8, 6))
plt.plot(x, y_sigmoid, label='Sigmoid Function', color='blue')
plt.plot(x, y_tanh, label='Hyperbolic Tangent (tanh)', color='red', linestyle='--')
plt.plot(x, y_softsign, label='Softsign Function', color='green', linestyle=':')
plt.title('Plot of Sigmoid, Tanh, and Softsign Functions')
plt.xlabel('x')
plt.ylabel('Function Value')
plt.grid(True)
plt.legend()
plt.show()
```

### **In paper2** we use **ReLU, Leaky ReLU and PReLU** activation functions:

### **ReLU (Rectified Linear Unit) function:**
$$f(x) = \max(0, x) \\
$$

### **PReLU (Parametric ReLU) function:**
$$ f(x) = \begin{cases}
 x & \text{if } x > 0 \\
 ax & \text{if } x \leq 0
 \end{cases}
 $$
 $$ $$

### **Leaky ReLU function:**
$$ f(x) = \begin{cases}
 x & \text{if } x > 0 \\
 ax & \text{if } x \leq 0
 \end{cases}$$



We update a in PRelu like this:

image.png

```python
# Define ReLU function
def relu(x):
    return np.maximum(0, x)

# Define PReLU function with parameter 'a'
def prelu(x, a):
    return np.where(x > 0, x, a * x)

# Define Leaky ReLU function with parameter 'a'
def leaky_relu(x, a):
    return np.where(x > 0, x, a * x)

# Generate input values
x = np.linspace(-10, 10, 100)

# Generate output values for each function
y_relu = relu(x)
y_prelu = prelu(x, 0.25)  # Assuming parameter 'a' as 0.1 for PReLU
y_leaky_relu = leaky_relu(x, 0.01)  # Assuming parameter 'a' as 0.01 for Leaky ReLU

# Plot ReLU, PReLU, and Leaky ReLU
plt.figure(figsize=(8, 6))
plt.plot(x, y_relu, label='ReLU', color='blue')
plt.plot(x, y_prelu, label='PReLU (a=0.25)', color='red', linestyle='--')
plt.plot(x, y_leaky_relu, label='Leaky ReLU (a=0.01)', color='green', linestyle='-.')
plt.title('Plot of ReLU, PReLU, and Leaky ReLU Functions')
plt.xlabel('x')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from tensorflow import keras
from keras import initializers

def build_model(input_shape, hidden_size, output_size, number_of_hidden_layers, activation_func, initializer):
  model = keras.Sequential()

  model.add(keras.Input(shape=input_shape))

  for _ in range(number_of_hidden_layers):
    model.add(keras.layers.Dense(hidden_size, activation=activation_func, kernel_initializer=initializer))
    model.add(keras.layers.Dropout(0.5))

  model.add(keras.layers.Dense(output_size, activation="softmax"))

  model.compile(optimizer="sgd", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
  return model
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def display_loss(history):
  plt.plot(history.history['loss'])
  plt.plot(history.history['val_loss'])
  plt.title('Model Loss')
  plt.ylabel('Loss')
  plt.xlabel('Number of Epochs')
  plt.legend(['Train', 'Test'], loc='upper left')
  plt.show()

def display_accuracy(history):
  plt.plot(history.history['accuracy'])
  plt.plot(history.history['val_accuracy'])
  plt.title('model accuracy')
  plt.ylabel('accuracy')
  plt.xlabel('no of epochs')
  plt.legend(['train', 'test'], loc='upper left')
  plt.show()
```

```python
import matplotlib.cm as cm  # Import colormap
def plot_activation_values(model, means_per_epoch, stds_per_epoch, percentiles_per_epoch):
  fig, ax1 = plt.subplots(figsize=(20, 6))
  num_dense_layers = sum(1 for layer in model.layers if isinstance(layer, keras.layers.Dense))

  colors = cm.rainbow(np.linspace(0, 1, 2 * num_dense_layers))
  skip_first = True

  for i, (layer, color) in enumerate(zip(model.layers, colors)):
    if(skip_first):
        skip_first = False
        continue
    if isinstance(layer, keras.layers.Dense):
        layer_means = [means_per_epoch[epoch][i//2] for epoch in range(epochs)]
        layer_stds = [stds_per_epoch[epoch][i//2] for epoch in range(epochs)]
        layer_percentiles = [percentiles_per_epoch[epoch][i//2] for epoch in range(epochs)]

        ax1.plot(range(1, epochs+1), layer_means, label=f'Mean (Layer {i+1})', color=color)
        ax1.errorbar(range(1, epochs+1), layer_means, yerr=layer_stds, fmt='.', color=color, label=f'Std Dev (Layer {i+1})')
        ax1.plot(range(1, epochs+1), layer_percentiles, '^', color = color, label=f'98th percentile (Layer {i+1})')

  ax1.set_xlabel('Epoch')
  ax1.set_ylabel('Activation Value')
  ax1.set_title('Activation Value Statistics')
  ax1.legend()

  plt.show()
```

```python

def train_model(model):
  losses = []
  val_losses = []
  accuracies = []
  val_accuracies = []
  means_per_epoch = []
  stds_per_epoch = []
  percentiles_per_epoch = []
  for epoch in range(epochs):
      train_data_gen = tensorflow.data.Dataset.from_tensor_slices((X_train, y_train)).batch(batch_size)
      layer_outputs = []
      for layer in model.layers:
          if isinstance(layer, keras.layers.Dense):
              layer_outputs.append([])
      for inputs, targets in train_data_gen:
          layer_input = inputs
          for i, layer in enumerate(model.layers):
              if isinstance(layer, keras.layers.Dense):
                  output = layer(layer_input)
                  layer_outputs[i//2].append(output.numpy().flatten())
                  layer_input = output
      means = []
      stds = []
      percentiles = []
      for activations_layer in layer_outputs:
            activations_layer = np.concatenate(activations_layer)
            means.append(np.mean(activations_layer))
            stds.append(np.std(activations_layer))
            percentiles.append(np.percentile(activations_layer, 98))
      means_per_epoch.append(means)
      stds_per_epoch.append(stds)
      percentiles_per_epoch.append(percentiles)
      print(f"Epoch: {epoch+1} / {epochs}")
      sigmoid_history_S = model.fit(X_train, y_train, batch_size=batch_size, epochs=1, validation_data=(X_val, y_val))
      losses.append(sigmoid_history_S.history['loss'])
      accuracies.append(sigmoid_history_S.history['accuracy'])
      val_losses.append(sigmoid_history_S.history['val_loss'])
      val_accuracies.append(sigmoid_history_S.history['val_accuracy'])
  return losses, val_losses, accuracies, val_accuracies, means_per_epoch, stds_per_epoch, percentiles_per_epoch
```

The bars representing the standard deviation for the first layer appear larger than the others because the activations in the first layer tend to have a larger variance compared to the subsequent layers.

This is a common phenomenon in neural networks, where the activations in the earlier layers (closer to the input) tend to have a higher variability or spread, while the activations in the later layers become more stable and have a smaller variance.

There are a few reasons why this happens:

1. **Input Distribution**: The activations in the first layer are directly influenced by the input data distribution, which can be quite diverse and have a large range of values, leading to higher variance in the activations.

2. **Weight Initialization**: The initial weights in the first layer are randomly initialized, which can introduce higher variability in the activations compared to later layers, where the weights have been optimized during training.

3. **Feature Extraction**: The early layers in a neural network are responsible for extracting low-level features from the input data, which can result in a wider range of activation values as the network learns to represent different patterns and features.

4. **Normalization**: Later layers in the network often benefit from the normalization or regularization techniques applied to the activations, such as batch normalization or dropout, which can reduce the variance and stabilize the activations.
